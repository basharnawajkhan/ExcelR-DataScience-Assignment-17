## Standardization

```python
std = fire4.values
scaler = StandardScaler().fit(std)
rescaledX = scaler.transform(std)

set_printoptions(precision=2)
print(rescaledX[0:5,:])
```

```python
X = std[:, 0:28]
Y = std[:, 28]
```

## Recursive Feature Elimination

```python
model = LogisticRegression()
rfe = RFE(model, 10)
fit = rfe.fit(X, Y)
```

```python
fit.n_features_
```

```python
fire5 = fire4.iloc[:,[4, 6, 9, 11, 12, 15, 17, 20, 21, 26, 28]]
fire5.head()
```

```python
array = fire5.values
array
```

## Train & Test Split

```python
x = array[:, 0:10]
y = array[:, 10]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True)
```

## Building model using Support Vector Machines

```python
model = SVC(kernel='rbf', random_state=7, gamma=50, C=5)
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy = accuracy_score(y_test, pred)
print((accuracy)*100)
```

```python
print("The accuracy of the svm classifier on training data is {:.2f} out of 1".format(model.score(x_train, y_train)*100))

print("The accuracy of the svm classifier on test data is {:.2f} out of 1".format(model.score(x_test, y_test)*100))
```
